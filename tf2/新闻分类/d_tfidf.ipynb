```python
from warnings import filterwarnings
filterwarnings('ignore')  # 不打印警告
from a_Data_preprocessing import load_xy
from b_metrics import metric, Timer
%matplotlib inline
```

## 数据加载、预处理

```python
from sklearn.feature_extraction.text import TfidfVectorizer
X_train, X_test, y_train, y_test = load_xy()
tfidf = TfidfVectorizer(token_pattern='\S+')
X_train = tfidf.fit_transform([' '.join(x) for x in X_train])
X_test = tfidf.transform([' '.join(x) for x in X_test])
```

## 模型评估函数

```python
def metrics(clf):
    t = Timer()  # 计时器
    y_pred = clf().fit(X_train, y_train).predict(X_test)
    metric(y_test, y_pred)
```

### 贝叶斯

```python
from sklearn.naive_bayes import MultinomialNB
metrics(MultinomialNB)
```

### 逻辑回归

```python
from sklearn.linear_model import LogisticRegression
metrics(LogisticRegression)
```

### 决策树

```python
from sklearn.tree import DecisionTreeClassifier
metrics(DecisionTreeClassifier)
```

### 随机森林

```python
from sklearn.ensemble import RandomForestClassifier
metrics(RandomForestClassifier)
```
